# Analysis and Findings

```python
cancellation_percentage = df['is_canceled'].value_counts(normalize=True)
print(cancellation_percentage)
```

In the above results, 0 represents the not-canceled ratio, while 1 represents the cancellation ratio.

# No-Cancellation Ratio vs Cancellation

```python
not_cancelled_percentage = cancellation_percentage[0] * 100
cancelled_percentage = cancellation_percentage[1] * 100
print(not_cancelled_percentage)
print(cancelled_percentage)
```

```python
print(f"No-Cancellation Ratio: {not_cancelled_percentage:.2f}%")
print(f"Cancellation Ratio: {cancelled_percentage:.2f}%")
```

```python
categories = ['No Cancellation Ratio', 'Cancellation Ratio']
cancel_ratio = [not_cancelled_percentage, cancelled_percentage]
colors = ('red','blue')
ax = sns.barplot(x = categories, y = cancel_ratio,palette=colors, width = 0.5)
plt.xlabel('Categories')
plt.ylabel('Percentages')
plt.title('Fig 1: No-Cancellation Ratio vs Cancellation Ratio')
plt.show()
```

# No-Cancellation and Cancellation Ratio in Each Hotel

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel_ratio = city_hotel['is_canceled'].value_counts(normalize = True)
print(city_hotel_ratio)
```

```python
city_hotel_cancellation_percentage = city_hotel_ratio[1] * 100
city_hotel_no_cancellation_percentage = city_hotel_ratio[0] * 100
print(city_hotel_cancellation_percentage)
print(city_hotel_no_cancellation_percentage)
```

```python
print(f"City Hotel Cancellation Percentage: {city_hotel_cancellation_percentage: .2f}%")
print(f"City Hotel No-Cancellation Percentage: {city_hotel_no_cancellation_percentage: .2f}%")
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel_ratio = resort_hotel['is_canceled'].value_counts(normalize = True)
print(resort_hotel_ratio)
```

```python
resort_hotel_cancellation_percentage = resort_hotel_ratio[0] * 100
resort_hotel_no_cancellation_percentage = resort_hotel_ratio[1] * 100
print(resort_hotel_cancellation_percentage)
print(resort_hotel_no_cancellation_percentage)
```

```python
print(f'Resort Hotel Cancellation Percentag: {resort_hotel_cancellation_percentage: .2f}%')
print(f'Resort Hotel No Cancellation Percentag: {resort_hotel_no_cancellation_percentage: .2f}%')
```

```python
ax = sns.countplot(x='hotel', hue='is_canceled', data=df, palette=['red',  'blue'], width=0.7)
ax.legend(labels=['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Fig 2: No-Cancellation and Cancellation Ratio in Each Hotel')
plt.xlabel('Hotels Names')
plt.ylabel('No-Cancellation / Cancellation Counts')
plt.show()
```

# City Hotel Cancellation Percentage VS No-Cancellation

```python
labels = ['Cancellation', 'No-Cancellation']
colors = ['red', 'green']
plt.pie([city_hotel_cancellation_percentage,
         city_hotel_no_cancellation_percentage],
        labels=labels, autopct='%1.2f%%', colors=colors)
plt.title('Fig 3: City Hotel Cancellation Percentage VS No-Cancellation')
plt.show()
```

# Resort Hotel Cancellation Percentage VS No-Cancellation

```python
labels = ['Cancellation', 'No-Cancellation']
colors = ['yellow', 'green']
plt.pie([resort_hotel_cancellation_percentage,
         resort_hotel_no_cancellation_percentage],
        labels=labels, autopct='%1.2f%%', colors=colors)
plt.title('Fig 4: Resort Hotel Cancellation Percentage VS No-Cancellation')
plt.show()
```

As we can see, resort hotels are experiencing a higher rate of reservation cancellations compared to city hotels

# Comparing Prices of Both Hotels

```python
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean() #adr = average daily rate
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.rc('font', size=18)

plt.figure(figsize=(16, 8))

plt.title('Fig 5: Average Daily Rate in City Hotel and Resort Hotel', fontsize=30, pad=30)
plt.xlabel('Months', labelpad=20)
plt.ylabel('Prices', labelpad=20)

plt.gca().set_facecolor('green')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel', color='red')
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel', color='yellow')
plt.legend(fontsize=16)
plt.show()
```

In the above plot, there are spikes. We can consider that, these spikes are on the weekends. On the weekends, the prices of City Hotel and Resort Hotels become high.

It also show that, in comparison to Resort Hotel, City Hotel have more bookings. It is possible that, Resort Hotels are more expensive than those in cities.


# Cancellation Rate on Monthly Basis

```python
monthName =df['arrival_date_month'].unique()
print(monthName)
```

```python
plt.rc('font', size=12)

plt.figure(figsize=(16, 8))
ax = sns.countplot(x=df['arrival_date_month'], hue='is_canceled', data=df, palette=['green', 'red'], width=0.7)
ax.legend(labels=['No-Cancellation', 'Cancellation'])
plt.title('Fig 6: Monthly Reservation Counts', fontsize=20, pad=20)
plt.xlabel('Months', labelpad=20)
plt.ylabel('Reservation Counts', labelpad=20)
ax.set_facecolor("grey")
plt.show()
```

Additionally, examining the average daily rate on a monthly basis can provide valuable insights.

```python
df['month'] = df['reservation_status_date'].dt.month
df['month'].unique()
```

```python
plt.rc('font', size=12)

data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

plt.figure(figsize=(15, 8))

ax = sns.barplot(x=monthName, y='adr', data=data,
            palette=['red', 'blue', 'green', 'purple', 'yellow', 'skyblue', 'grey'])
plt.title("Fig 7: Monthly Average Daily Rate", fontsize=20, pad=20)
plt.xlabel('Months', labelpad=20)
plt.ylabel('Average Daily Rate', labelpad=20)
ax.set_facecolor("white")
plt.show()
```
